```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
(iris_x, iris_y) = load_iris(return_X_y = True)
train_x, test_x, train_y, test_y = train_test_split(iris_x, iris_y)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(8, 8, 8), activation="relu", solver="lbfgs")
mlp.fit(train_x, train_y)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(mlp.predict(iris_x), iris_y))
```

```python
def to_fixed(n):
    return round(n * 2048)
```

```python
import struct
def output_row(r):
    x = [to_fixed(n) for n in r]
    s = b''.join(struct.pack("hh", int(x[i+1]) or 0, int(x[i])) for i in range(0, len(x), 2)).hex()
    s = s + "0" * (16 * 4 - len(s))
    out = ", ".join(["0x" + s[i:i+8] for i in range(0, len(s), 8)])
    return out
```

```python
for layer in range(len(mlp.coefs_)):
    print("l{}:".format(layer))
    for row in range(mlp.coefs_[layer].shape[1]):
        print("dat " + output_row(mlp.coefs_[layer][: , row]))
```

```python
for row in range(len(iris_x)):
    print(output_row(iris_x[row]))
```
